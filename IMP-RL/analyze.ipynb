```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

# 加载 .p 文件
file_path = 'rl_results/imp_job_222/test_data/test_resultauto_data.p'
with open(file_path, 'rb') as f:
    data = pickle.load(f)

exps = data[0][2]  # 示例：第一个模型的第一个游戏

# 绘制 Exploitability 随迭代次数变化的图
plt.plot(np.arange(len(exps)), exps, marker='o')
plt.title('Exploitability over PSRO Iterations')
plt.xlabel('Iteration')
plt.ylabel('Exploitability')
plt.grid(True)
plt.show()

print(data)
```
